```python
Rate_merged_Q22025 = Rate_merged.replace(',','', regex=True)
```

```python
Rate_merged_Q22025['Total DL Amount'] = pd.to_numeric(Rate_merged['Total DL Amount'])
Rate_merged_Q22025['Dimensional Weight'] = pd.to_numeric(Rate_merged['Dimensional Weight'])
Rate_merged_Q22025['Gross Weight_y'] = pd.to_numeric(Rate_merged['Gross Weight_y'])
```

```python
Rate_merged_Q22025.info()
```

```python
# calculations for per price average of each zip code.
avg_dw = Rate_merged_Q22025.groupby('Destination Location Zip Code')['Dimensional Weight'].mean()
dim_wgt = (avg_dw*.99969093)
dim_wgt_df = pd.DataFrame(dim_wgt).reset_index()
dim_wgt_df
```

```python
# sorting Total DL column by greater than 0
avg_dl = Rate_merged[Rate_merged["Total DL Amount"] > 0]

# grouping by dest zip and getting average of the Total DL column per zip
avg_dl_cost = avg_dl.groupby('Destination Location Zip Code')['Total DL Amount'].mean()

# Creating DataFrame
avg_dl_cost_df = pd.DataFrame(avg_dl_cost).reset_index()

# creating variable to hold dimensional weight
dim_wgt_2 = dim_wgt_df['Dimensional Weight']
actual_wgt = Rate_merged['Gross Weight_y']

# creating two new rows in dataframe and formatting.
avg_dl_cost_df["AVG Dim Wght per Zip"] = dim_wgt_2.map("{:.0f}".format)
avg_dl_cost_df["AVG Dim/LB rate per Zip"] = (avg_dl_cost_df["Total DL Amount"]/dim_wgt_2).map("$ {:.2f}".format)
avg_dl_cost_df["AVG Actual Wght per Zip"] = actual_wgt.map("{:.0f}".format)
avg_dl_cost_df["AVG Act/LB rate per Zip"] = (avg_dl_cost_df["Total DL Amount"]/actual_wgt).map("$ {:.2f}".format)


# renaming Total DL Amount column
avg_dl_cost_df = avg_dl_cost_df.rename(columns={"Total DL Amount": "AVG Delivery Rate per Zip"})
avg_dl_cost_df['AVG Delivery Rate per Zip'] = avg_dl_cost_df['AVG Delivery Rate per Zip'].map("$ {:.2f}".format)
avg_dl_cost_df
```

```python
Airport_df = Airport_df.dropna(subset=['Destination Location Zip Code'])
Airport_df
```

```python
avg_dl_cost_df2 = avg_dl_cost_df.merge(Airport_df,how = "left", on=['Destination Location Zip Code'] )
avg_dl_cost_df2
```

```python
avg_dl_cost_df2.to_csv("Q4 Average Delivery Cost per Zip.csv")
```
